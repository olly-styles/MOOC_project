```python
import pandas as pd
import numpy as np
import pickle 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

df_all = pd.read_pickle('../data/df_combined.pkl')
df_lecv = df_all[[0,1,2,3,4,5]]
df_lecd = df_all[[6,7,8,9,10,11]]
df_quiz = df_all[[12,13,14,15,16,17]]
df_comments = df_all[[18,19,20,21,22,23]]
df_threads = df_all[[24,25,26,27,28,29]]
df_posts = df_all[[30,31,32,33,34,35]]
df_grades = pd.DataFrame(df_all['grade'])

def create_violin(df,drop,y_axis,featurenum,ylim,scale):
    stacked = df.stack()
    stacked = stacked.reset_index()
    stacked = pd.DataFrame(stacked)
    stacked = stacked.set_index('anon_user_id')
    stacked = stacked.merge(df_grades, left_index=True,right_index=True)
    stacked = stacked.rename(columns={0: y_axis,'level_1':'week'})
    stacked.ix[stacked['week'] == featurenum + '_w1','week'] = 1
    stacked.ix[stacked['week'] == featurenum + '_w2','week'] = 2
    stacked.ix[stacked['week'] == featurenum + '_w3','week'] = 3
    stacked.ix[stacked['week'] == featurenum + '_w4','week'] = 4
    stacked.ix[stacked['week'] == featurenum + '_w5','week'] = 5
    stacked.ix[stacked['week'] == featurenum + '_w6','week'] = 6

    stacked = stacked[stacked['grade'] != drop]
    
    fig, ax = plt.subplots(figsize=(20,6))

    
    sns.violinplot(x="week", y=y_axis, hue="grade", data=stacked,split=True,cut=0\
               ,palette="Set2",scale_hue=True,scale=scale).set(ylim=(0, ylim));
    
    sns.set(font_scale=2.5) 
```

```python
create_violin(df_lecv,'fail','Lecture views','01',30,'width')
create_violin(df_lecd,'fail','Lecture downloads','02',30,'width')
create_violin(df_quiz,'fail','Quiz attempts','03',6,'width')
create_violin(df_comments,'fail','Forum comments','04',1,'width')
create_violin(df_threads,'fail','Forum threads','05',1,'width')
create_violin(df_posts,'fail','Forum posts','06',1,'width')
```

```python
create_violin(df_lecv,'pass','Lecture views','01',30,'width')
create_violin(df_lecd,'pass','Lecture downloads','02',30,'width')
create_violin(df_quiz,'pass','Quiz attempts','03',6,'width')
create_violin(df_comments,'pass','Forum comments','04',2,'width')
create_violin(df_threads,'pass','Forum threads','05',2,'width')
create_violin(df_posts,'pass','Forum posts','06',2,'width')
```
